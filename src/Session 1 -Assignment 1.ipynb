#### 2. Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included). The numbers obtained should be printed in a comma-separated sequence on a single line.
    

```python
numbers = []
for n in range (2000, 3201):
    if n%7 == 0 and n%5 !=0:
        numbers.append(n)
#print('List of numbers divisible by 7, but not a multiple of 5 are below. \n{}'.format(numbers)) 

print(*numbers, sep=',')        
```

#### 4.Write a Python program to find the volume of a sphere with diameter 12 cm. 
   **`Formula: V=4/3 * Ï€ * r 3`**

```python
PI = 3.14
diameter = 12
r = diameter/2
volume = 4/3 * PI * (r ** 3)

print ("Volume of sphere having diameter {}cm is {}cm3.".format(diameter, round(volume,2)))
```

#### 1. Write a program which accepts a sequence of comma-separated numbers from console and generate a list.

```python
numbers = input("Enter comma separated numbers:")
numbers_list = []
for n in numbers.split(','):
    if n.isnumeric():
        numbers_list.append(int(n))
        
print('Numbers as list\n',numbers_list)
```

#### 2. Create the below pattern using nested for loop in Python.
`*`
<br>`**`
<br>`***` 
<br>`****`
<br>`*****`
<br>`****`
<br>`***`
<br>`**`
<br>`*`


```python
c = '*'
n = 10
m = n//2 + 1
for i in range(n):
    if i < m:
        for j in range(i):
            print(c, end = '')
            j = j + 1
        print("")    
    else: 
        for j in range(n, i, -1):
            print(c, end = '')
            j = j - 1     
        print("")    
        
```

```python
# Another way of doing it for my understanding
c = '*'
n = 10
m = n//2 + 1
for i in range(n):
    if i < m:
        print(c*i)   
    else: 
         print(c*(n-i))  
```
